```python
#1: IMPORTS

from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.callbacks import EarlyStopping

# seasonal decomposition (used once, no data mutation)
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#2: LOAD AND INSPECT DATA

df = pd.read_csv("merged_potato_reservoir.csv")
df.head()# STEP 2: LOAD AND INSPECT DATA

import pandas as pd

# Load dataset
df = pd.read_csv("merged_potato_reservoir.csv")

# Preview first few rows
print(" Dataset Preview:")
display(df.head())

# Shape of dataset
print(f"\nDataset shape: {df.shape[0]} rows × {df.shape[1]} columns")

# Data types
print("\n Column Data Types:")
print(df.dtypes)

# Missing values
print("\n Missing Values per Column:")
print(df.isnull().sum())
```

```python
#3: SORT & DROP DATE
df = df.copy()
df['temperature_recorded_date'] = pd.to_datetime(df['temperature_recorded_date'], errors='coerce')
df.sort_values(by=['temperature_recorded_date', 'state_name'], inplace=True)
df.drop(columns=['temperature_recorded_date'], inplace=True)

print(" Sorted by date/state and dropped 'temperature_recorded_date'.")
display(df.head())
```

```python
#4: COLUMN-WISE INTERPOLATION (LINEAR)
for col in df.columns:
    if df[col].dtype != 'object':
        df[col] = df[col].interpolate(method='linear', limit_direction='both')

print("Linear interpolation done for numeric columns.")
print("\nRemaining missing values:")
print(df.isnull().sum())
```

```python
#5: AUTOENCODER-BASED ANOMALY REMOVAL (PER STATE)
df_auto = df.copy()

# numeric features for AE
num_cols = df_auto.select_dtypes(include=[np.number]).columns.tolist()
if 'yield' in num_cols:   # exclude target if present
    num_cols.remove('yield')

states = df_auto['state_name'].dropna().unique().tolist() if 'state_name' in df_auto.columns else ['All']

cleaned_parts = []
total_removed = 0

for state in states:
    part = df_auto[df_auto['state_name'] == state].copy() if state != 'All' else df_auto.copy()

    # skip AE if too small
    if len(num_cols) < 1 or part.shape[0] < 50:
        cleaned_parts.append(part)
        continue

    # local scaling for AE
    scaler_local = MinMaxScaler()
    X = scaler_local.fit_transform(part[num_cols])

    # tiny AE (adam + mse), bottleneck ~ half
    input_dim = X.shape[1]
    inp = Input(shape=(input_dim,))
    encoded = Dense(max(2, input_dim // 2), activation='relu')(inp)
    decoded = Dense(input_dim, activation='sigmoid')(encoded)
    ae = Model(inp, decoded)
    ae.compile(optimizer='adam', loss='mse')

    es = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)
    ae.fit(X, X, epochs=100, batch_size=32, verbose=0, callbacks=[es])

    # reconstruction error
    X_hat = ae.predict(X, verbose=0)
    mse = np.mean((X - X_hat) ** 2, axis=1)

    # threshold = mean + 3*std
    th = mse.mean() + 3*mse.std()
    keep_mask = mse <= th

    removed = int((~keep_mask).sum())
    total_removed += removed

    cleaned_parts.append(part.loc[keep_mask].copy())

df_ae = pd.concat(cleaned_parts, axis=0).sort_index()
print(f" AE anomaly removal done. Rows removed: {total_removed}")
display(df_ae.head())
    
```

```python
#6: SEASONAL DECOMPOSITION
try:
    num_cols_post = df_ae.select_dtypes(include=[np.number]).columns.tolist()
    if num_cols_post:
        series_for_decomp = df_ae[num_cols_post[0]].reset_index(drop=True)
        if series_for_decomp.dropna().shape[0] > 120:
            _ = seasonal_decompose(series_for_decomp.interpolate(), model='additive', period=100)
    print(" seasonal_decompose executed")
except Exception as e:
    print("seasonal_decompose skipped:", e)
```

```python
#7: GLOBAL MINMAX SCALER (0,1)
df_scaled = df_ae.copy()
num_cols_global = df_scaled.select_dtypes(include=[np.number]).columns.tolist()
if num_cols_global:
    scaler_global = MinMaxScaler(feature_range=(0, 1))
    df_scaled[num_cols_global] = scaler_global.fit_transform(df_scaled[num_cols_global])

print(" MinMax scaling done (0–1) on numeric columns.")
display(df_scaled.head())
```

```python
#8: ONE-HOT ENCODE 'state_name' 

from sklearn.preprocessing import OneHotEncoder

if 'state_name' in df_scaled.columns:
    ohe = OneHotEncoder(handle_unknown='ignore')
    state_ohe = ohe.fit_transform(df_scaled[['state_name']]).toarray()  # ensure dense array always
    ohe_cols = [f"state_name_{c}" for c in ohe.categories_[0].tolist()]
    df_ohe = pd.DataFrame(state_ohe, columns=ohe_cols)
    df_ohe.index = df_scaled.index  # align index before concat

    df_final = pd.concat([df_scaled.drop(columns=['state_name']), df_ohe], axis=1)
else:
    df_final = df_scaled.copy()

print(" One-Hot Encoding for 'state_name' completed successfully.")
display(df_final.head())
```

```python
#9: SAVE CLEANED DATASET
out_path = "potato_reservoir_cleaned_dataset.csv"
df_final.to_csv(out_path, index=False)

print(" Cleaning complete.")
print(f"Saved cleaned dataset to: {out_path}")
print("Final shape:", df_final.shape)
print("\nRemaining missing values after all steps:")
print(df_final.isnull().sum())
```

```python
# DETECT TARGET & PREPARE COLUMNS (use df_scaled with state_name)

from sklearn.model_selection import train_test_split

# confirm target
if 'yield' not in df_scaled.columns:
    print("Columns:", list(df_scaled.columns))
    raise ValueError("Target column 'yield' not found. Tell me the correct target name to proceed.")

target_col = 'yield'
feature_cols = [c for c in df_scaled.columns if c not in [target_col]]  # keep state_name for filtering
print(" Target:", target_col)
print(" Features (count):", len(feature_cols))
```

```python
# TRAIN/TEST SPLIT 

states = df_scaled['state_name'].dropna().unique().tolist()
print(f"States found: {len(states)} ->", states)

X_train_list, X_test_list, y_train_list, y_test_list = [], [], [], []

for st in states:
    part = df_scaled[df_scaled['state_name'] == st].copy()

    X_part = part[feature_cols].drop(columns=['state_name'], errors='ignore')
    y_part = part[target_col]

    if len(part) < 5:  # avoid tiny states breaking the split
        print(f"Skipping state '{st}' (too few rows: {len(part)})")
        continue

    X_tr, X_te, y_tr, y_te = train_test_split(
        X_part, y_part, test_size=0.20, random_state=42
    )

    X_train_list.append(X_tr)
    X_test_list.append(X_te)
    y_train_list.append(y_tr)
    y_test_list.append(y_te)
    
X_train = pd.concat(X_train_list, axis=0).reset_index(drop=True)
X_test  = pd.concat(X_test_list, axis=0).reset_index(drop=True)
y_train = pd.concat(y_train_list, axis=0).reset_index(drop=True)
y_test  = pd.concat(y_test_list, axis=0).reset_index(drop=True)

print(" State-wise split completed successfully!")
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
#  Scatter Plots

import matplotlib.pyplot as plt
import numpy as np

def plot_reg_scatter(y_true, y_pred, title):
    plt.figure(figsize=(6.5,6.5))
    
    # Scatter points
    plt.scatter(y_true, y_pred,
                color='#4A90E2',   # light blue
                edgecolor='none',
                alpha=0.6,
                s=45)
    
    # Regression line (best linear fit)
    coeffs = np.polyfit(y_true, y_pred, 1)
    line = np.poly1d(coeffs)
    x_vals = np.linspace(y_true.min(), y_true.max(), 100)
    plt.plot(x_vals, line(x_vals),
             color='red', linewidth=2)
    
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xlabel("Actual Yield", fontsize=12)
    plt.ylabel("Predicted Yield", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.35)
    plt.show()

# RF
plot_reg_scatter(y_test, y_pred, "Random Forest: Predicted vs Actual")

# XGBoost
plot_reg_scatter(y_test, y_pred_xgb, "XGBoost: Predicted vs Actual")

# Ensemble (XGBoost + Random Forest)
plot_reg_scatter(y_test, y_pred_ensemble, "Ensemble (XGB + RF): Predicted vs Actual")

# Optimized RF
plot_reg_scatter(y_test, y_pred_best, "Optimized RF: Predicted vs Actual")
```

```python
# Residual Distribution (per model)

import numpy as np
import matplotlib.pyplot as plt

def plot_residuals(y_true, y_pred, title):
    residuals = y_pred - y_true
    mu, sigma = float(np.mean(residuals)), float(np.std(residuals))

    plt.figure(figsize=(7,5))
    plt.hist(residuals, bins=40, edgecolor='black', alpha=0.75)
    plt.title(f"{title}\nResidual Distribution (μ={mu:.4f}, σ={sigma:.4f})", fontsize=14)
    plt.xlabel("Residual")
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.35)
    plt.show()

    print(f"{title} -> mean(residual)={mu:.6f}, std(residual)={sigma:.6f}")

# Random Forest
plot_residuals(y_test, y_pred, "Random Forest")

# XGBoost
plot_residuals(y_test, y_pred_xgb, "XGBoost")

# Ensemble (XGB + RF)
plot_residuals(y_test, y_pred_ensemble, "Ensemble (XGB + RF)")

# Optimized Random Forest
plot_residuals(y_test, y_pred_best, "Optimized RF (GridSearchCV)")
```

```python
# Residual-Based Accuracy Calculation 

import numpy as np
import pandas as pd

# Calculate residuals
res_rf  = y_test - y_pred
res_xgb = y_test - y_pred_xgb

# Define acceptable error thresholds (absolute residual range)
thresholds = [0.001, 0.005, 0.01, 0.02]  # = 0.1%, 0.5%, 1%, 2%

# Function to calculate % of predictions within threshold
def residual_accuracy(y_true, y_pred, thresholds):
    res = np.abs(y_true - y_pred)
    results = []
    for t in thresholds:
        acc = np.mean(res <= t) * 100
        results.append({"Residual ≤": f"±{t}", "Accuracy (%)": acc})
    return pd.DataFrame(results)

# Compute for both models
rf_acc_table  = residual_accuracy(y_test, y_pred, thresholds)
xgb_acc_table = residual_accuracy(y_test, y_pred_xgb, thresholds)

print(" Residual-Based Accuracy (Random Forest):")
display(rf_acc_table)

print(" Residual-Based Accuracy (XGBoost):")
display(xgb_acc_table)
```

```python
# V2: prepare dataframe + recover `state_name` if OHE exists

import pandas as pd
import numpy as np

assert 'df_final' in globals() or 'df_scaled' in globals(), "Run cleaning cells first."

if 'df_final' in globals():
    df_v2 = df_final.copy()
else:
    df_v2 = df_scaled.copy()

state_ohe_cols = [c for c in df_v2.columns if c.startswith('state_name_')]

if state_ohe_cols:
    st_series = df_v2[state_ohe_cols].idxmax(axis=1).str.replace('state_name_', '', regex=False)
    df_v2['__state_name__'] = st_series
else:
    if 'state_name' not in df_v2.columns:
        raise ValueError("No state info found. Provide df_final with state_name_* OHE or df_scaled with state_name.")
    df_v2['__state_name__'] = df_v2['state_name']

print(" Version-2 (V2) dataframe ready.")
print("Detected state indicator columns:", state_ohe_cols[:5], "..." if len(state_ohe_cols)>5 else "")
print("Shape:", df_v2.shape)
```

```python
# state-wise chronological split 

from math import floor

target_col = 'yield'

numeric_cols = [c for c in df_v2.columns if df_v2[c].dtype != 'object']
if target_col not in numeric_cols:
    raise ValueError("Target 'yield' must be numeric (it was during cleaning).")

df_tmp = df_v2.copy()
df_tmp['__row_order__'] = np.arange(len(df_tmp))

train_idx, test_idx = [], []

for st in df_tmp['__state_name__'].unique():
    block = df_tmp[df_tmp['__state_name__'] == st].sort_values('__row_order__')
    n = len(block)
    k = floor(0.8 * n)  
    train_idx.extend(block.index[:k])
    test_idx.extend(block.index[k:])

X_cols_all = [c for c in numeric_cols if c not in [target_col]]  
X_cols_all = [c for c in X_cols_all if c != 'crop_name']

X_train_v2 = df_tmp.loc[train_idx, X_cols_all].copy()
y_train_v2 = df_tmp.loc[train_idx, target_col].copy()
X_test_v2  = df_tmp.loc[test_idx,  X_cols_all].copy()
y_test_v2  = df_tmp.loc[test_idx,  target_col].copy()

print(" V2 state-wise split complete (within each state).")
print("Train shape:", X_train_v2.shape, "| Test shape:", X_test_v2.shape)
print("States used:", df_tmp['__state_name__'].nunique())
```

```python
# RandomForest training + evaluation 

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

print(" Training RandomForestRegressor (V2 state-wise split)...")
rf_v2 = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)
rf_v2.fit(X_train_v2, y_train_v2)
print(" Training completed!")

# predictions
y_pred_v2_rf = rf_v2.predict(X_test_v2)

# metrics
mse_v2_rf  = mean_squared_error(y_test_v2, y_pred_v2_rf)
mae_v2_rf  = mean_absolute_error(y_test_v2, y_pred_v2_rf)
rmse_v2_rf = np.sqrt(mse_v2_rf)
r2_v2_rf   = r2_score(y_test_v2, y_pred_v2_rf)

print("\n Evaluation Metrics — Random Forest (V2 temporal split)")
print(f"MSE  : {mse_v2_rf:.6f}")
print(f"MAE  : {mae_v2_rf:.6f}")
print(f"RMSE : {rmse_v2_rf:.6f}")
print(f"R²   : {r2_v2_rf:.6f}")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np

X_train_v2 = df_tmp.loc[train_idx, X_cols_all].values
y_train_v2 = df_tmp.loc[train_idx, target_col].values.reshape(-1, 1)
X_test_v2  = df_tmp.loc[test_idx,  X_cols_all].values
y_test_v2  = df_tmp.loc[test_idx,  target_col].values.reshape(-1, 1)

x_scaler = StandardScaler()
y_scaler = MinMaxScaler(feature_range=(0, 1))

X_train_s = x_scaler.fit_transform(X_train_v2)
X_test_s  = x_scaler.transform(X_test_v2)
y_train_s = y_scaler.fit_transform(y_train_v2).ravel()
y_test_s  = y_scaler.transform(y_test_v2).ravel()
```

```python
def create_sequences(X, y, window):
    X_seq, y_seq = [], []
    for i in range(len(X) - window):
        X_seq.append(X[i:i+window])
        y_seq.append(y[i+window])
    return np.array(X_seq, np.float32), np.array(y_seq, np.float32)

WINDOW = 24  # can tune later

X_train_seq, y_train_seq = create_sequences(X_train_s, y_train_s, WINDOW)
X_test_seq,  y_test_seq  = create_sequences(X_test_s,  y_test_s, WINDOW)

print("Train seq:", X_train_seq.shape, "| Test seq:", X_test_seq.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, Add, Attention
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

def lstm_v2_block(input_shape):
    inputs = Input(shape=input_shape)

    l1 = LSTM(128, return_sequences=True)(inputs)
    l2 = LSTM(128, return_sequences=True)(l1)
    res = Add()([l1, l2])

    attn = Attention(use_scale=True)([res, res])
    final_lstm = LSTM(32)(attn)

    drop = Dropout(0.3)(final_lstm)
    output = Dense(1, activation='linear')(drop)

    return Model(inputs, output)

model_lstm_v2 = lstm_v2_block((X_train_seq.shape[1], X_train_seq.shape[2]))
model_lstm_v2.compile(optimizer="adam", loss="mse")

es = EarlyStopping(monitor="val_loss", patience=6, restore_best_weights=True)
rlr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-5, verbose=1)

history = model_lstm_v2.fit(
    X_train_seq, y_train_seq,
    validation_split=0.1,
    epochs=50,
    batch_size=64,
    verbose=1,
    callbacks=[es, rlr]
)

print(" LSTM  Training Completed!")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_pred_s = model_lstm_v2.predict(X_test_seq, verbose=0).ravel()

y_pred = y_scaler.inverse_transform(y_pred_s.reshape(-1,1)).ravel()
y_true = y_scaler.inverse_transform(y_test_seq.reshape(-1,1)).ravel()

mse_lstm_v2  = mean_squared_error(y_true, y_pred)
mae_lstm_v2  = mean_absolute_error(y_true, y_pred)
rmse_lstm_v2 = np.sqrt(mse_lstm_v2)
r2_lstm_v2   = r2_score(y_true, y_pred)

print("\n LSTM Results")
print(f"MSE  : {mse_lstm_v2:.6f}")
print(f"MAE  : {mae_lstm_v2:.6f}")
print(f"RMSE : {rmse_lstm_v2:.6f}")
print(f"R²   : {r2_lstm_v2:.6f}")
```

```python
# Build supervised sequences for LSTM (per state, chronological)

import numpy as np

LOOKBACK = 12  

def build_sequences(df, state_col="__state_name__", feature_cols=None, target_col="yield", lookback=12):
    X_seq, y_seq, idx_seq = [], [], []

    for st in df[state_col].unique():
        block = df[df[state_col] == st].sort_values("__row_order__")
        Xb = block[feature_cols].values
        yb = block[target_col].values
        ib = block.index.values  

        if len(block) <= lookback:
            continue

        for t in range(lookback, len(block)):
            X_seq.append(Xb[t-lookback:t, :])  
            y_seq.append(yb[t])               
            idx_seq.append(ib[t])             

    X_seq = np.asarray(X_seq, dtype=np.float32)
    y_seq = np.asarray(y_seq, dtype=np.float32)
    idx_seq = np.asarray(idx_seq)
    return X_seq, y_seq, idx_seq

X_seq_all, y_seq_all, idx_seq_all = build_sequences(
    df_tmp, state_col="__state_name__", feature_cols=X_cols_all, target_col=target_col, lookback=LOOKBACK
)

print(" Sequences built.")
print("X_seq_all:", X_seq_all.shape, "(samples, lookback, features)")
print("y_seq_all:", y_seq_all.shape)
```

```python
# sequences include past 'yield' (lags), lookback=24

import numpy as np

LOOKBACK = 24  


feature_cols_plus = X_cols_all + (['yield'] if 'yield' not in X_cols_all else [])

def build_sequences_with_yield(df, state_col="__state_name__", feature_cols=None, target_col="yield", lookback=24):
    X_seq, y_seq, idx_seq = [], [], []
    for st in df[state_col].unique():
        block = df[df[state_col] == st].sort_values("__row_order__")
        Xb = block[feature_cols].values
        yb = block[target_col].values
        ib = block.index.values
        if len(block) <= lookback:
            continue
        for t in range(lookback, len(block)):
            # window t-lookback...t-1 contains past yields & features; target is y at t
            X_seq.append(Xb[t-lookback:t, :])
            y_seq.append(yb[t])
            idx_seq.append(ib[t])
    return np.asarray(X_seq, np.float32), np.asarray(y_seq, np.float32), np.asarray(idx_seq)

X_seq_all, y_seq_all, idx_seq_all = build_sequences_with_yield(
    df_tmp, state_col="__state_name__", feature_cols=feature_cols_plus, target_col=target_col, lookback=LOOKBACK
)

train_mask = np.isin(idx_seq_all, np.array(train_idx))
test_mask  = np.isin(idx_seq_all, np.array(test_idx))
X_train_seq, y_train_seq = X_seq_all[train_mask], y_seq_all[train_mask]
X_test_seq,  y_test_seq  = X_seq_all[test_mask],  y_seq_all[test_mask]

print(" Sequences rebuilt with past yield.")
print("X_train_seq:", X_train_seq.shape, "| X_test_seq:", X_test_seq.shape)
```

```python
# Split sequences into V2 train/test using target indices

train_mask = np.isin(idx_seq_all, np.array(train_idx))
test_mask  = np.isin(idx_seq_all, np.array(test_idx))

X_train_seq, y_train_seq = X_seq_all[train_mask], y_seq_all[train_mask]
X_test_seq,  y_test_seq  = X_seq_all[test_mask],  y_seq_all[test_mask]

print(" BiLSTM V2 sequence split complete.")
print("X_train_seq:", X_train_seq.shape, "| y_train_seq:", y_train_seq.shape)
print("X_test_seq :", X_test_seq.shape,  "| y_test_seq :", y_test_seq.shape)
```

```python
# BiLSTM model 

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Bidirectional, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

print(" BiLSTM training started...")

model_lstm = Sequential([
    Bidirectional(LSTM(64, return_sequences=True), input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
    Dropout(0.3),
    LSTM(32, return_sequences=False),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1)
])

model_lstm.compile(optimizer="adam", loss="mse")

es = EarlyStopping(monitor="val_loss", patience=6, restore_best_weights=True)
rlr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-5, verbose=1)

history = model_lstm.fit(
    X_train_seq, y_train_seq,
    validation_split=0.1,  
    epochs=60,
    batch_size=64,
    verbose=1,
    callbacks=[es, rlr]
)

print(" BiLSTM training completed!")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

y_pred_bi_s = model_lstm.predict(X_test_seq, verbose=0).ravel()

y_pred_bilstm_v2 = y_scaler.inverse_transform(y_pred_bi_s.reshape(-1,1)).ravel()
y_true_v2        = y_scaler.inverse_transform(y_test_seq.reshape(-1,1)).ravel()

mse_bilstm_v2  = mean_squared_error(y_true_v2, y_pred_bilstm_v2)
mae_bilstm_v2  = mean_absolute_error(y_true_v2, y_pred_bilstm_v2)
rmse_bilstm_v2 = np.sqrt(mse_bilstm_v2)
r2_bilstm_v2   = r2_score(y_true_v2, y_pred_bilstm_v2)

print("\n BiLSTM Results (V2)")
print(f"MSE  : {mse_bilstm_v2:.6f}")
print(f"MAE  : {mae_bilstm_v2:.6f}")
print(f"RMSE : {rmse_bilstm_v2:.6f}")
print(f"R²   : {r2_bilstm_v2:.6f}")
```

```python
# Forecast
MODEL = xgb_v2         # change to rf_v2 if needed
MODEL_NAME = "XGBoost (V2)"   # update if switching model
TARGET_YEAR = 2022

# Last K rows used for evaluation
K = 100

# Feature Columns for V2 Predictions
state_cols = [c for c in df_final.columns if c.startswith("state_name_")]
base_cols = [
    'apy_item_interval_start','state_temperature_max_val',
    'state_temperature_min_val','state_rainfall_val',
    'FRL','Live Cap FRL','Level','Current Live Storage'
]
feature_cols_v2 = base_cols + state_cols

# Collect results
results = []
for st_col in state_cols:
    st_name = st_col.replace("state_name_", "")
    df_state = df_final[df_final[st_col] == 1].copy()

    if df_state.empty:
        continue

    df_state = df_state.sort_values("__row_order__", ascending=True)

    y_true = df_state['yield'].values[-K:]
    X_vals = df_state[feature_cols_v2].values[-K:]
    y_pred = MODEL.predict(X_vals)

    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)

    results.append({
        "state_name": st_name,
        "year": TARGET_YEAR,
        "y_true_mean": round(y_true.mean(), 6),
        "y_pred_mean": round(y_pred.mean(), 6),
        "mse": mse,
        "rmse": rmse,
        "mae": mae
    })

# Put into DataFrame
results_df = pd.DataFrame(results)

# Display Output Cleanly
print(f"\n Model Used: {MODEL_NAME}")
print(f" State-wise Forecast Results for YEAR {TARGET_YEAR}")
display(results_df.round(6))
```
